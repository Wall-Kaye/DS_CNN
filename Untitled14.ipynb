```python
import numpy as np
import keras
import time
import os
import cv2
from keras import backend as k
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense, Flatten, Dropout
from keras.layers import Conv2D, MaxPooling2D
#from keras.layers.normalization import BatchNormalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    BatchNormalization, SeparableConv2D, MaxPooling2D, Activation, Flatten, Dropout, Dense
)

from tensorflow.keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt 
```

```python
train_path= 'TpDevSurv//train'#chemin vers les donées d'entrainement
batch_size=25#taille du lot.On envoie les images par lot de 25 afin d'aller vite.Par exemple si on choisi
#d'envoyer les images une par une ça n'a pas beaucoup d'interets
valid_path='TpDevSurv//valid'
train_batches =ImageDataGenerator().flow_from_directory(train_path, target_size=(80,80),batch_size=batch_size)
valid_batches =ImageDataGenerator().flow_from_directory(valid_path, target_size=(80,80),batch_size=batch_size)
```

```python
steps_train=train_batches.n/batch_size#le nombre totales d'images d'entrainenement divisé par le lots d'entrainement
#dans notre cas de lots de 25 images
print(steps_train)
steps_valid=valid_batches.n/batch_size
print(steps_valid)
num_classes=train_batches.num_classes#le nombre total de classes dans le lot d'entrainement
print(num_classes)
cls_train = train_batches.classes
cls_test =  valid_batches.classes
class_names = list(train_batches.class_indices.keys())
print(class_names)
```

```python

def path_join(dirname, filenames):
    return [os.path.join(dirname, filename) for filename in filenames]
image_paths_train = path_join(train_path, train_batches.filenames)
image_paths_test = path_join(valid_path, valid_batches.filenames)

def plot_images(images, cls_true, cls_pred=None, smooth=True):
    assert len(images) == len(cls_true)

    # creation de la figure dans lequel on va afficher nos elements
    fig, axes = plt.subplots(3, 3)

    # quelques ajustements lors de l'affichage
    if cls_pred is None:
        hspace = 0.3
    else:
        hspace = 0.6
    fig.subplots_adjust(hspace=hspace, wspace=0.3)

    if smooth:
        interpolation = 'spline16'
    else:
        interpolation = 'nearest'

    for i, ax in enumerate(axes.flat):
        
        if i < len(images):
            ax.imshow(images[i],
                      interpolation=interpolation)

            cls_true_name = class_names[cls_true[i]]

            if cls_pred is None:
                xlabel = "True: {0}".format(cls_true_name)
            else:
                cls_pred_name = class_names[cls_pred[i]]
                xlabel = "True: {0}\nPred: {1}".format(cls_true_name, cls_pred_name)
            ax.set_xlabel(xlabel)
        ax.set_xticks([])
        ax.set_yticks([])
    
    plt.show()

def load_images(image_paths):
    images = [plt.imread(path) for path in image_paths]
    return np.asarray(images)
images = load_images(image_paths=image_paths_train[0:9])
cls_true = cls_train[0:9]
plot_images(images=images, cls_true=cls_true, smooth=True)
```

```python
model = Sequential()#ça va nous servir de conteneur
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu', input_shape=(80,80,3)))
#128 le nombre de filtre de sortie de la convolution
#kernel_size=(3,3) determine la hauteur et la largeur de la fenêtre de convolution
#activation=relu c'est le nom de la fonction d'activation
#input_shape=(100,100,3) la forme d'entrée de mes images,on a la largeur,la hauteur et le canal(rgb) d'où 3
"""
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())#nomalisation des données
model.add(MaxPooling2D((2, 2), strides=(2, 2)))"""

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())#nomalisation des données
model.add(MaxPooling2D((2, 2), strides=(2, 2))) #on fait le maxpooling.
#cette instruction permet de faire le maximum de pooling 

model.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
model.add(Conv2D(16, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
#La normalisation par lots applique une transformation qui maintient la sortie moyenne
#proche de 0 et l’écart type de sortie proche de 1.
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(Conv2D(8, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
#La normalisation par lots applique une transformation qui maintient la sortie moyenne
#proche de 0 et l’écart type de sortie proche de 1.
model.add(MaxPooling2D((2, 2), strides=(2, 2)))
model.add(Flatten())#construction d'un vecteur avec la matrice afin d'être evoyer au reseau dens
model.add(Dropout(0.5))
model.add(Dense(512,activation='relu'))
#the classifier
model.add(Dense(num_classes, activation='softmax'))
#model.add(Dense(num_classes, activation='sigmoid'))
#model dense pour classer les resultats.Softmax pour classer par probabilité
#nombre d'objets à classer et la fonction d'activation softmax
#softmax pour avoir la probabilité entre nos differentes classes

model.summary() 
```

```python
epochs=40
#compile the model
model.compile(Adam(),loss='categorical_crossentropy', metrics=['accuracy'])
#adam pour optimiser l'apprentissage

start = time.time()#debut de la compilation
history = model.fit_generator(generator=train_batches,
                                  epochs=epochs,
                                  steps_per_epoch=steps_train,
                                  validation_data=valid_batches,
                                  validation_steps=steps_valid)
end = time.time()#fin de la compilation

print ("Model took %0.2f seconds to train"%(end - start))#on affiche la difference de temps d'apprentissage
result = model.evaluate_generator(valid_batches, steps=1)
print("Test-set classification accuracy: {0:.2%}".format(result[1]))
```

```python
def plot_model_history(model_history):
    fig, axs = plt.subplots(1,2,figsize=(15,5))
    # summarize history for accuracy
    axs[0].plot(range(1,len(model_history.history['accuracy'])+1),model_history.history['accuracy'])
    axs[0].plot(range(1,len(model_history.history['val_accuracy'])+1),model_history.history['val_accuracy'])
    axs[0].set_title('Model Accuracy')
    axs[0].set_ylabel('Accuracy')
    axs[0].set_xlabel('Epoch')
    axs[0].set_xticks(np.arange(1,len(model_history.history['accuracy'])+1))
    axs[0].legend(['train', 'val'], loc='best')
    # summarize history for loss
    axs[1].plot(range(1,len(model_history.history['loss'])+1),model_history.history['loss'])
    axs[1].plot(range(1,len(model_history.history['val_loss'])+1),model_history.history['val_loss'])
    axs[1].set_title('Model Loss')
    axs[1].set_ylabel('Loss')
    axs[1].set_xlabel('Epoch')
    axs[1].set_xticks(np.arange(1,len(model_history.history['loss'])+1))
    axs[1].legend(['train', 'val'], loc='best')
    plt.show()

plot_model_history(history)
```
